```python
# 2. Listor (din watchlist)
watchlist = ["ERIC-B.ST", "VOLV-B.ST", "TSLA", "SWED-A.ST", "HM-B.ST", "ATCO-A.ST"]
print("Antal aktier i listan:", len(watchlist))
print("FÃ¶rsta aktien:", watchlist[0])
print("Tre fÃ¶rsta:", watchlist[:3])
```

```python
# 3. For-loop â€“ kolla hela listan
for aktie in watchlist:
    print("Laddar data fÃ¶r", aktie)
```

```python
# 4. Importera yfinance och hÃ¤mta data fÃ¶r alla pÃ¥ en gÃ¥ng
import yfinance as yf
import pandas as pd

data = yf.download(watchlist, start="2025-01-01", end="2025-11-30")['Adj Close']
print(data.tail(7))  # sista 7 dagarna
```

```python
# 5. RÃ¤kna ut veckans procentfÃ¶rÃ¤ndring fÃ¶r alla aktier
returns = data.pct_change().iloc[-1] * 100
print("\nVeckans fÃ¶rÃ¤ndring (%):")
print(returns.sort_values(ascending=False).round(2))
```

```python
# 6. Vem gick bÃ¤st i veckan?
bÃ¤st = returns.idxmax()
bÃ¤st_procent = returns.max()
print(f"\nDagens vinnare: {bÃ¤st} +{bÃ¤st_procent:.2f}% ðŸ”¥")
```

```python
# 7. Plotta alla aktier i en snygg graf
import matplotlib.pyplot as plt

plt.figure(figsize=(14,8))
for aktie in watchlist[:6]:  # plottar bara 6 stycken sÃ¥ det inte blir kaos
    plt.plot(data.index, data[aktie], label=aktie.replace(".ST",""))

plt.title("Min svenska watchlist 2024â€“2025", fontsize=18)
plt.xlabel("Datum")
plt.ylabel("Pris (SEK)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
