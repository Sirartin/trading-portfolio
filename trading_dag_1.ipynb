```python
# Ladda hem Apple-aktien och rita en graf – bara för att det är coolt
import yfinance as yf
import matplotlib.pyplot as plt

# Hämta data
data = yf.download("AAPL", start="2023-01-01", end="2025-11-24")
print(data.tail())

# Rita slutkurs
plt.figure(figsize=(12,6))
plt.plot(data.index, data['Close'], label="Apple", color="#00aa00", linewidth=2)
plt.title("AAPL – min första trading-graf!", fontsize=16)
plt.xlabel("Datum")
plt.ylabel("Pris (USD)")
plt.grid(True)
plt.legend()
plt.show()
```
